# **Make a Scatterplot Game**

Randomly generate points on a graph and the player has to input the (x,y) coordinates.


```python
import matplotlib.pyplot as plt
import random

xmin = -20
xmax = 20
ymin = -20
ymax = 20

fig, ax = plt.subplots()

xpoint = random.randint(xmin,xmax)
ypoint = random.randint(ymin,ymax)
x = [xpoint]
y = [ypoint]
ax.set_title("Scatter Plot Game")
plt.axis([xmin,xmax,ymin,ymax])
plt.plot([xmin,xmax],[0,0],'b')
plt.plot([0,0],[ymin,ymax],'b')
plt.plot(x,y,'ro')
print(" ")
plt.grid()
plt.show()
print(" ")
guess = input("Enter the coordinates of the red point: \n")
guess_array = guess.split(",")
xguess = int(guess_array[0])
yguess = int(guess_array[1])

# Display Successful or unsuccessful message
if xguess == xpoint and yguess == ypoint:
  print("You guessed it correctly.  The red point is at ({0},{1}).  Nice job!".format(x[0],y[0]))
else:
  print("You missed the red point.")
```

# **Algebra Practice Game**

Generate one step and two step problems with random integer values and the player has to input the answer.

```python
import sympy
from sympy import symbols
from sympy.solvers import solve
import random

def algebra():
  eq = None
  x = symbols('x')
  generator = random.randint(1,100)
  if generator >= 50:
    num = random.randint(1,100)
    if generator <=75:
      print("Equation: y = x - {0}".format(num))
      eq = x - num
    else:
      print("Equation: y = x + {0}".format(num))
      eq = x + num
  else:
    num1 = random.randint(1,100)
    num2 = random.randint(1,100)
    if generator <=12:
      print("Equation: y = {0} * x + {1}".format(num1,num2))
      eq = num1*x + num2
    elif generator > 12 and generator <=24:
      print("Equation: y = -{0} * x + {1}".format(num1,num2))
      eq = -num1*x + num2
    elif generator > 24 and generator <=36:
      print("Equation: y = {0} * x - {1}".format(num1,num2))
      eq = num1*x - num2
    else:
      print("Equation: y = -{0} * x - {1}".format(num1,num2))
      eq = -num1*x - num2

  answer = input("Solve for X: ")
  if "/" in answer:
    nd = answer.split("/")
    n = float(nd[0])
    d = float(nd[1])
    answer = n/d
  else:
    answer = float(answer)
  solution = solve(eq,x)
  sol = str(solution[0])
  if "/" in sol:
    nd = sol.split("/")
    n = float(nd[0])
    d = float(nd[1])
    sol = n/d
  else:
    sol = float(sol)

  # Display successful or unsuccessful message
  if sol == answer:
    print("Correct!")
  else:
    print("Incorrect.  Answer is {0}".format(solution[0]))

algebra()
```

# **Projectile Game**

Display a wall with random height and location


```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import *
import random

def projectile():
  targetx = None
  targety = None
  wallx = random.randint(1,18)
  wally = random.randint(1,18)
  def f(a,b,c):
    xmin = 0
    xmax = 20
    ymin = 0
    ymax = 20
    points = 10*(xmax-xmin)
    x = np.linspace(xmin,xmax,points)
    plt.axis([xmin,xmax,ymin,ymax])
    plt.plot([wallx,wallx],[0,wally],'r')
    targetx = wallx
    targety = wally + 1
    plt.plot([targetx],[targety],'go')
    y = a*x**2 + b*x + c
    plt.plot(x,y)
    plt.show()
    return targetx, targety

  tracker = True
  print("Try to enter the correct a,b and c values to get the vertex on the green dot!")
  while tracker is True:
    print("Enter 999 in a to exit the program if needed!")
    a = int(input("a value: "))
    b = int(input("b value: "))
    c = int(input("c value: "))
    if a == 999:
      tracker = False
    vx = -b/(2*a)
    vy = a*vx**2 + b*vx + c
    targetx, targety = f(a,b,c)
    print("Target vertex: ( {0}, {1} )".format(targetx,targety))
    print("Projection vertex: ( {0}, {1} )".format(vx,vy))

    # Display Successful message
    if vx == targetx and vy == targety:
      tracker = False
      print("You got it.  Great job!")

projectile()
```
